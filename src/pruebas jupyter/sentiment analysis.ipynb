```python
import nltk 
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import requests
import pandas as pd
```

```python

sid = SentimentIntensityAnalyzer()

sid.polarity_scores("Weekend proyect")
```

```python
url='http://localhost:8080/'
chats = requests.get(url).json()
chats
```

```python
chats_df = pd.DataFrame(chats)
chats_df
chats_df
```

```python
chats_df.to_json('chatsfelipe.json')
```

```python
texts = chats_df['text']
lista = []

for e in texts:
    lista.append("{}: {}".format(e, sid.polarity_scores(e)))

lista
```
